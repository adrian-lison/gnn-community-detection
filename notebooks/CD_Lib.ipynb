```python
def number_of_clusters(com):
    number_of_clusters = 0
    for x in com.communities:
        number_of_clusters += 1
    return(number_of_clusters)
```

```python
def com_to_labels(com):
    i = 0
    pred_labels = [None]*g.number_of_nodes()
    for x in com.communities:
        for y in x:
            pred_labels[y] = i
        i = i + 1
    return(pred_labels)        
```

```python
import performance as pf
```

```python
def get_info(com):
    print(number_of_clusters(com))
    print(pf.rand_score(labels,com_to_labels(com)))
    print(pf.mutual_info_score(labels,com_to_labels(com)))
    print(pf.variation_of_information_score(labels,com_to_labels(com)))
```
